```python
def printinfo():  # define or created function called print info for printing the information.
    temp = pd.DataFrame(index=df.columns)
    temp['data_type'] = df.dtypes #(This returns a Series with the data type of each column)
    temp['null_count'] = df.isnull().sum()
    temp['unique_count'] = df.nunique() #(returns the number of unique values for each column)
    return temp
```

```python
# Data Modeling
data_model_x=df[['Category','Reviews','Size','Installs','Type','Price','Content Rating','Genres','Last Updated']]
data_model_y=df[['Rating']]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(data_model_x[['Last Updated']])
data_model_x[['Last Updated']]=scaler.transform(data_model_x[['Last Updated']])

data_model_x
```

```python
encoded_x=pd.get_dummies(data_model_x, columns=['Category',"Content Rating","Type","Genres"])
#It converts categorical data into dummy or indicator variables.
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(encoded_x,data_model_y,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
cls=LinearRegression()

cls.fit(x_train,y_train)
cls.predict(x_test)
```

```python
cls.score(x_train,y_train)
```

```python
cls.score(x_test,y_test)
```
